```python
# Question 5

# set up inputs for M&M probabilities by color
brown = 0.12
yellow = 0.15
red = 0.12
blue = 0.23
orange = 0.23
green = 0.15

# Compute the probability that a randomly selected M&M is not blue
Pr_Not_Blue = 1 - blue
print(f"The probability that a randomly selected M&M is not blue is {Pr_Not_Blue}." )


# Compute the probability that a randomly selected M&M is brown or green
Pr_Brown_Or_Green = brown + green
print(f"The probability that a randomly selected M&M is brown or green is {Pr_Brown_Or_Green}.")

# Compute the probability that two randomly selected M&Ms are both green
Pr_Two_Greens = green * green
print(f"The probability that two randomly selected M&Ms are both green is {Pr_Two_Greens}.")


# If you randomly select 6 M&M compute the probability that none of them are orange
Pr_No_Oranges = (1 - orange) ** 6
print(f"The probability that none of the randomly selected M&Ms are orange is {Pr_No_Oranges}.")


# If you randomly select 6 M&Ms, compute the probability that at least one of them is orange

# The probability that at least one M&M is orange equals 1 minus the probability of zero oranges
Pr_At_Least_One_Orange = 1 - Pr_No_Oranges
print(f"The probability of at least one randomly selected M&M being orange is {Pr_At_Least_One_Orange}.")



```

#### Answer

```python
# Question 7

# First we will define our inputs

# Proportion of all drivers
prop_teen = 0.12
prop_ya = 0.24 # young adule
prop_ma = 0.48 # middle aged
prop_senior = 0.16

# Probability of at least one collision
teen_collision = 0.14
ya_collision = 0.06
ma_collision = 0.05
senior_collision = 0.06

# Find the probability that a driver with at least one collision in the past year is a teen driver

# Pr_teen_given_collision = pr_teen_and_collision / pr_collision

pr_teen_and_collision = prop_teen * teen_collision

pr_collision = (prop_teen * teen_collision) + (prop_ya * ya_collision) + (prop_ma * ma_collision) + (prop_senior * senior_collision)

pr_teen_given_collision = pr_teen_and_collision / pr_collision

print(f"The probability that a driver with at least one collision is a teen is {pr_teen_given_collision}.")
```

### Question 10

#### Question

```python
plt.figure(figsize = (13, 13))
question_10 = plt.imread('HW_Assignment_4_Question_10.jpg')
plt.imshow(question_10)
plt.axis("off")
plt.show()
```

#### Answer

```python
# Question 10

pr_seatbelt = 0.64

pr_ejected_given_seatbelt = 0.16

pr_ejected_given_no_seatbelt = 0.46

# Find the probability that a randomly selected person who was ejected was wearing a seatbelt

# pr(seatbelt | ejected ) = pr(seatbelt & ejected) / pr(ejected)
pr_seatbelt_and_ejected = pr_ejected_given_seatbelt * pr_seatbelt
pr_ejected = (pr_ejected_given_seatbelt * pr_seatbelt) + (pr_ejected_given_no_seatbelt * (1 - pr_seatbelt))
pr_seatbelt_given_ejected = pr_seatbelt_and_ejected / pr_ejected
print(f"The probability that a randomly selected person who was ejected was wearing a seatbelt is {pr_seatbelt_given_ejected}.")

# Find the probability that a randomly selected person who was ejected was not wearing a seatbelt

# pr(no seatbelt | not ejected ) = pr(not ejected & no seatbelt ) / pr(not ejected)
pr_no_seatbelt_and_not_ejected = (1 - pr_seatbelt) * (1 - pr_ejected_given_no_seatbelt)
pr_not_ejected = pr_no_seatbelt_and_not_ejected + (pr_seatbelt * (1 - pr_ejected_given_seatbelt))
pr_no_seatbelt_given_not_ejected = pr_no_seatbelt_and_not_ejected / pr_not_ejected
print(f"The probability that a person was not wearing a seatbelt given that they are not ejected is {pr_no_seatbelt_given_not_ejected}.")
```
