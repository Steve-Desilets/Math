```python
# Question 4

# Let's use matrix multiplication to calculate the cost of each ingredient needed for one batch of candy when purchasing
# ingredients from supplier K

import numpy

Quantity_Matrix = [[6, 8, 1],
    [6, 4, 1],
    [5, 7, 1]]
Quantity_Matrix = matrix(Quantity_Matrix)

Price_Matrix = [3, 5, 2]
Price_Matrix = transpose(matrix(Price_Matrix))

Total_Cost_Matrix = 100*numpy.dot(Quantity_Matrix, Price_Matrix)

print(f"The cost for 100 batches of cherry candy ingredients (when purchasing from supplier K) is ${Total_Cost_Matrix[0]}.")
print(f"The cost for 100 batches of almond candy ingredients (when purchasing from supplier K) is ${Total_Cost_Matrix[1]}.")
print(f"The cost for 100 batches of raisin candy ingredients (when purchasing from supplier K) is ${Total_Cost_Matrix[2]}.")

print(f"The total cost of all the ingredients for 100 batches of each candy (when purchasing from supplier K) is ${numpy.sum(Total_Cost_Matrix):.2f}.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (15, 15))
question_1 = plt.imread('Midterm_Exam_Question_6.jpg')
plt.imshow(question_1)
plt.axis("off")
plt.show()
```

#### Answer

```python
# Question 6

# First let's solve this problem via the Simplex Method

# Import necessary methods and functions
from pulp import LpVariable, LpProblem, LpMaximize, LpStatus, value, LpMinimize

# Create linear programming variables 
x = LpVariable("x", 0, None) # x >= 0 & no upper
y = LpVariable("y", 0, None) # y >= 0 & no upper

# Create the problem variable
prob = LpProblem("problem", LpMaximize)

# Define the constraints - remember that the lower bounds are defined in LpVariable above
prob += 6*x + 2*y <= 18000
prob += 3*x + 4*y <= 12000

# Define the objective function to maximize
prob+= 0.25*x + 0.16*y

# Solve the problem
status = prob.solve()
LpStatus[status]
maximum_profit = (0.25 * value(x)) + (0.16 * value(y))


print(f"The maximum profit occurs when the company produces {value(x)} liters of chemicals via the old process and {value(y)} liters of chemicals via the new process.")
print(f"The maximum profit at these production levels is ${maximum_profit}.")



# Second, let's check our work by solving graphically / geometrically


# Import needed libraries
import matplotlib.pyplot
import numpy

# Create figure and define our equations
matplotlib.pyplot.figure()
n = numpy.arange(0, 10000, 0.01)
r1 = (-3) * n + 9000
r2 = (-0.75)*n + 3000

# Set parameters for the graph
matplotlib.pyplot.xlim(0, 10000)
matplotlib.pyplot.ylim(0, 10000)
matplotlib.pyplot.hlines(0, 0, 10000, color = 'k')
matplotlib.pyplot.vlines(0, 0, 10000, color = 'k')
matplotlib.pyplot.grid(True)

matplotlib.pyplot.xlabel('Liters Produced Via The Old Process')
matplotlib.pyplot.ylabel('Liters Produced Via The New Process')
matplotlib.pyplot.title('Liters of Chemicals Produced')

# Plot the lines that determine our feasible region
matplotlib.pyplot.plot(n, r1, color = 'b')
matplotlib.pyplot.plot(n, r2, color = 'r')


# Solve for the corners of our feasible region

# Bottom Left Corner
a1 = numpy.array([[1,0], [0,1]])
b1 = numpy.array([0,0])
c1 = numpy.linalg.solve(a1,b1)

# Top Left Corner
a2 = numpy.array([[1,0], [3,4]])
b2 = numpy.array([0, 12000])
c2 = numpy.linalg.solve(a2, b2)

# Top Right Corner
a3 = numpy.array([[6,2], [3,4]])
b3 = numpy.array([18000, 12000])
c3 = numpy.linalg.solve(a3, b3)

# Bottom Right Corner
a4 = numpy.array([[0,1], [6,2]])
b4 = numpy.array([0, 18000])
c4 = numpy.linalg.solve(a4, b4)

#Create the shaded region
x = [c1[0], c2[0], c3[0], c4[0]]
y = [c1[1], c2[1], c3[1], c4[1]]
matplotlib.pyplot.fill(x,y,color = "yellow")

matplotlib.pyplot.show()

# Figure out which corner of the feasible region maximizes the revenue function

def profit_calculator(a, b):
    profit = 0.25*a + 0.16*b
    print(f"The profit achieved when {a} liters of chemicals are produced via the old process and {b} liters of chemicals are produced via the new process is ${profit:.2f}.")

profit_calculator(c1[0], c1[1])
profit_calculator(c2[0], c2[1])
profit_calculator(c3[0], c3[1])
profit_calculator(c4[0], c4[1])

```

```python
 # Question 7
    
# First, let's solve this problem using the simplex method

# Import necessary methods and functions
from pulp import LpVariable, LpProblem, LpMaximize, LpStatus, value, LpMinimize

# Create linear programming variables
t = LpVariable("t", 1, None) # t >= 0 & no upper
a = LpVariable("a", 10, None) # a >= 0 & no upper

# Create the problem variable
prob = LpProblem("problem", LpMinimize)

# Define the constraints - remember that the lower bounds are defined in LpVariable above
prob += t + a <= 45
prob += t + a >= 30
prob += 3*t >= 2*a


# Define the objective function to maximize
prob+= 2400*t + 1100*a

# Solve the problem
status = prob.solve()
LpStatus[status]
minimum_cost = 2400*value(t) + 1100*value(a)

print(f"The minimum monthly cost is ${minimum_cost:.2f} and occurs when Northwestern employs {value(t):.0f} teachers and {value(a):.0f} TAs per month.")
```

#### Answer

```python
# Question 9

# Let's solve this problem via the Simplex Method

# Import necessary methods and functions
from pulp import LpVariable, LpProblem, LpMaximize, LpStatus, value, LpMinimize

# Create linear programming variables 
d = LpVariable("d", 0, None) # d >= 0 & no upper
s = LpVariable("s", 0, None) # s >= 0 & no upper
t = LpVariable("t", 30, None) # t >= 30 & no upper

# Create the problem variable
prob = LpProblem("problem", LpMaximize)

# Define the constraints - remember that the lower bounds are defined in LpVariable above
prob += d + s + t <= 210
prob += -d + 2*s <= 0

# Define the objective function to maximize
prob+= 200*d + 1000*s + 300*t

# Solve the problem
status = prob.solve()
LpStatus[status]
maximum_revenue = (200 * value(d)) + (1000 * value(s)) + (300 * value(t))


print(f"The maximum revenue occurs when the company stocks {value(d):.0f} DVD players, {value(s):.0f} speaker systems, and {value(t):.0f} televisions.")
print(f"The revenue at these storage levels is ${maximum_revenue:.2f}.")
```

```python
# Question 10

# Let's apply Bayes' Theorem to solve for the probability that a ticket came from Box A given that it's a winning ticket

pr_a = 0.25

pr_win_given_a = 0.02

pr_b = 1 - pr_a

pr_win_given_b = 0.05

pr_a_given_win = ( (pr_win_given_a * pr_a ) / ( (pr_win_given_a * pr_a) + (pr_win_given_b * pr_b) ) )

print(f"The probability that a ticket came from box A given that it's a winning ticket is {pr_a_given_win}.")
```
