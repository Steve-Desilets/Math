```python
# Question 4

# First we need to solve for the function f(x) = -5x^2 + 50x + 3 at the points x = 1 and x = 9

from sympy import Symbol, lambdify, solve, solveset, Interval

x = Symbol('x')
f = -5*x**2 + 50*x + 3
f_function = lambdify(x, f)

x1 = 1
x2 = 9

y1 = f_function(x1)
y2 = f_function(x2)

print(f"The first coordinate of interest is ({x1},{y1}).")
print(f"The second coordinate of interest is ({x2},{y2}).")

# We see that f(1) = f(9).  We can also see that this function will be differentiable 
# over the domain [1,9].  Therefore, Rolle's Theorem applies here.  

# Accordingly we must find where f'(x) = 0 in the domain [1,9]

from sympy import plotting

f_derivative = f.diff(x)

derivative_roots = solveset(f_derivative, x, Interval(1,9)) # Find the roots of our derivative 

print(f"The slope of the function is zero at x = {derivative_roots}.")

plotting.plot(f, xlim = (0, 10), ylim = (-200, 300)) # plot the function for visual inspection
```

```python
# Question 5

# First we want to find the average slope of the function f(x) = 6*sqrt(x) + 6 on the 
# interval of x: [2,6].

from sympy import Symbol, solveset, solve, Interval, diff, sqrt

x = Symbol('x')

f_x = 6*sqrt(x) + 6

f_x_function = lambdify(x, f_x)

x1 = 2
x2 = 6

y1 = f_x_function(2)
y2 = f_x_function(6)

average_slope = (y2 - y1) / (x2 - x1)

print(f"The average slope of f(x) over the interval [2,6] is {average_slope}.")

# Second, we want to find the value of c such that f'(c) = this mean slope (within the domain
# of x that is specified above).

f_x_derivative = f_x.diff(x)

c = solveset(f_x_derivative - average_slope, x, Interval(2,6)) # Find where the slope equals the average slope 

print(f"The slope of the function equals the average slope for the domain at x = {c}.")

plotting.plot(f_x, xlim = (1, 7)) # plot the function for visual inspection
```

```python
# Question 8

# Consider the function f(x) 4x + 9/x

# Find the x-values of the critical points (A, B, and C)

print(f"B is 0 since the function is undefined there.")

from sympy import solveset, Interval, oo, Symbol, diff, lambdify

x = Symbol('x')
f = 4*x + (9/x)

f_deriv = f.diff(x)

critical_x_vals = solveset( f_deriv, x, Interval(-oo,oo))

print(f"The critical values for x are {critical_x_vals}.")

# Determine whether f(x) is increasing or decreasing between each critical point

f_deriv_func = lambdify(x, f_deriv)

def slope_function(x):
    if f_deriv_func(x) > 0:
        print(f"The slope is increasing at x = {x}.")
    elif f_deriv_func(x) < 0:
        print(f"The slope is decreasing at x = {x}.")
    elif f_deriv_func(x) == 0:
        print(f"The slope is zero at x = {x}.")
    else:
        print(f"The slope is undefined at x = {x}.")

slope_function(-5)
slope_function(-1)
slope_function(1)
slope_function(5)
```

```python
# Question 14

# Monthly sales of a product are modeled by S = 180*t / ( (t+ 3)**2) for t>=0

# where is is in thousands of dollars and t is in months.  After how many months will sales be 
# maximized and what will the maximum sales be?

from sympy import Symbol, solveset, plotting, oo, Interval

t = Symbol('t')
S = 180*t / ( (t+ 3)**2)
S_deriv = S.diff(t)


print(S_deriv)

critical_t_values = solveset( S_deriv, t, Interval(0, oo))

S_func = lambdify(t, S)

critical_S_values = [S_func(t) for t in critical_t_values]

print(f"The critical points occur at t equals {critical_t_values} and S equals {critical_S_values}, respectively.")

plotting.plot(S, xlim = (0, 20), ylim =(-1, 20) )
```
