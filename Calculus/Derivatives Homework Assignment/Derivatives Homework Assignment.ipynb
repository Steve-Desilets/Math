```python
# Question 2

# We need to find the limit of f(x) as x approaches 2 for the function:
# (x^3 + x^2 - 11x + 10) / (x -2)


# first let's see whether we can factor the numerator to see if we can elicit its behavior
# as x approaches 2 by canceling out x - 2 with the x - 2 in the denominator
from sympy import expand, symbols, factor
  
x = symbols('x')
polynomial = x**3 + x**2 - 11*x + 10
  
# Use sympy.factor() method
fact = factor(polynomial)
  
print(fact)

# We see that (x-2) is a factor of the numerator
# Let's evaluate the rest of the numerator to see what it approaches as x approaches 2

answer_q1 = 2**2 + 3*2 - 5
print(f"As x approaches 2, this equation should approach {answer_q1}.")

# Let's check our work by observing the graph of this function

from sympy import Symbol, limit, lambdify
x=Symbol('x')
y = (x**3 + x**2 -11*x +10)/(x-2)

from numpy import linspace
import matplotlib.pyplot as plt

x_plot = linspace(-3,5,100)  
f_plot = lambdify(x,y) 
y_plot = f_plot(x_plot) 

plt.figure()
plt.plot(x_plot,y_plot) 
plt.title('Graph of x and f(x)')
plt.ylabel('(x^3 + x^2 -11x +10)/(x-2)')
plt.xlabel('x')
plt.grid()
plt.show()

# Let's do one last check by plugging in values that are close to x = 2

def q2_calculator(x):
    f_x = (x**3 + x**2 -11*x +10)/(x-2)
    print(f"The value of f(x) at {x} is {f_x}.")
    
q2_calculator(1.9)
q2_calculator(1.99)
q2_calculator(2.01)
q2_calculator(2.1)
```

```python
# Question 5

# For part A, please note that we are finding the average velocity - not the instantaneous velocity.  
# Therefore, taking the derivative will not be helpful
# Instead we need to use the position function, which we'll use in the calculations below


def average_velocity_calculator(time_start, time_end):
    height_start = (48*time_start) - (16*(time_start**2))
    height_end = (48*time_end) - (16*(time_end**2))
    average_velocity = (height_end - height_start) / (time_end - time_start)
    print(f"The average velocity between {time_start} and {time_end} is {average_velocity}.")
    
average_velocity_calculator(3, 3.01)
average_velocity_calculator(3, 3.005)
average_velocity_calculator(3, 3.002)
average_velocity_calculator(3, 3.001)

# For part B, we do need to calculate the instantaneous velocity so we will take the derivative for that

from sympy import Symbol, diff, lambdify

from IPython.display import display

t=Symbol('t')
p=(48*t) - (16*(t**2))
der_f = p.diff(t)
func_der_f = lambdify(t, der_f)

display(der_f)
print(f"The rate of change at t=3 is {func_der_f(3)}.")
```

```python
# Question 8

# We are given that the height of a swallow at time t is h(t) = -4t^2 + 3t + 62

# Find the height above ground of the swallow before taking flight 
def swallow_height_calculator(time):
    height = (-4*(time**2)) + 3*time + 62
    print(f"The height of the swallow at time {time} seconds is {height} meters.")

swallow_height_calculator(0)
    
# Find the height above the ground of the swallow at t = 3.32 seconds
swallow_height_calculator(3.32)


# Find the initial velocity of the swallow
from sympy import Symbol, diff, lambdify

from IPython.display import display

t=Symbol('t')
p = (-4*(t**2)) + 3*t + 62
der_f = p.diff(t)
func_der_f = lambdify(t, der_f)

display(der_f)

print(f"The velocity of the swallow at time = 0 seconds is {func_der_f(0)} meters per second.")


# Find the velocity of the swallow at t = 3.32 seconds
print(f"The velocity of the swallow at time = 3.32 seconds is {func_der_f(3.32)} meters per second.")


# Find the initial acceleration of the swallow
from sympy import Symbol, diff, lambdify

from IPython.display import display

t=Symbol('t')
p = (-4*(t**2)) + 3*t + 62
second_der_f = diff(p, t, 2)
func_second_der_f = lambdify(t, second_der_f)

display(second_der_f)
print(f"The acceleration of the swallow at time = 0 seconds is {func_second_der_f(0)} meters per second squared.")


# Find the acceleration of the swallow at t = 3.32 seconds
print(f"The acceleration of the swallow at time = 3.32 seconds is {func_second_der_f(3.32)} meters per second squared.")
```

```python
# Question 11 - Attempt 1

# We are given the formula for the labor cost to produce yachts as 
# f(x) = (x^2 + 6x + 4)^2

# Find the marginal cost to produce the 10th yacht
def marginal_cost_calculator(n):
    cost_n_minus_one = ((n-1)**2 + 6*(n-1) + 4)**2
    cost_n = (n**2 + 6*n + 4)**2
    marginal_cost = cost_n - cost_n_minus_one
    print(f"The marginal cost to produce the {n}th yacht is ${marginal_cost}.")
    
marginal_cost_calculator(10)

# Find the marginal cost to produce the 40th yacht
marginal_cost_calculator(40)
```

```python
# Question 15

# The depth (in feet) of water at a dock changes with the rise and fall of tides.
# The depth is modeled by the function:
# D(t) = 5*sin( (pi/3)*t + (8*pi/3) ) + 4
#where t is the number of hours after midnight.  Find the rate at which the depth is changing at 3am.


from sympy import Symbol, diff, lambdify, sin, pi

from IPython.display import display

t = Symbol('t')
D = 5*sin( (pi*t/3) + (8*pi/3)  ) + 4
der_D = D.diff(t)
display(der_D)

func_der_D = lambdify(t, der_D)
print(f"The rate fo change of the water depth at 3am is {func_der_D(3)} feet per hour.")
```
