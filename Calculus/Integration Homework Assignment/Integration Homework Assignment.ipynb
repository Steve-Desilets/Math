```python
# Question 5

# You are memorizing words for a vocabulary test.  You studied a few days
# ago, and you know 18 words already.  It is the night before the test, and 
# you sit down to finish studying the words.  Suppose that the rate of 
# memorizing is given by M'(t) = -0.007t^2 + 0.2t where M'(t) is the memory
# rate, in words per minute.  What is the total number of words you have 
# memorized after 8 minutes of studying

from sympy import Symbol, integrate
t = Symbol('t')
M_prime = -0.007*t**2 + 0.2*t
M_t = integrate(M_prime, t) + 18
words_memorized = M_t.evalf(subs={t:8})
print(f"After 8 minutes of studying, I have memorized {words_memorized} words.")
```

```python
# Question 11

# The lifetime of a printer costing $120 is exponentially distributed
# with mean 3.98 years.  The manufacturer agrees to pay a full refund
# to a buyer if the printer fails during the first half year following 
# its purchase and a one-half refund if it fails during the second half 
# year.

# If the manufacturer sells 350 printers, how much should it expect to pay
# in refunds?


from sympy import Symbol, integrate, exp
t = Symbol('t')
lamb = 1/2.18
t_prime = lamb * exp(-t * lamb)
refunds_first_half_year = 200 * 250 * integrate(t_prime, (t, 0, 0.5))
refunds_second_half_year = 100 * 250 * integrate(t_prime, (t, 0.5, 1))
total_refunds = refunds_first_half_year + refunds_second_half_year
print(f"The manufacturer should expect to pay ${total_refunds:.2f} in refunds.")
```

```python
# Question 14

# The heights of adult men in America are normally distributed, with a 
# mean of 69.1 inches and a standard deviation of 2.69 inches.  The 
# heights of adult women in America are also normally distributed, but 
# with a mean of 64.6 inches and a standard deviation of 2.52 inches.


# If a man is 6 feet 3 inches tall, what is his z-score (to two decimal
# places)?

ZScore_14A =  ((6*12 + 3) - 69.1) / 2.69
print(f"If a man is 6 feet 3 inches tall, then his Z-Score is {ZScore_14A}.")


# What percentage of men are shorter than 6 feet 3 inches?  Round to
# the nearest tenth of a percent.

from scipy.stats import norm
answer_14B = norm(loc = 69.1, scale = 2.69).cdf(12*6 + 3)
print(f"The proportion of men shorter than 6 feet 3 inches is {answer_14B}.")

# If a woman is 5 feet 11 inches tall, what is her z-score
# (to two decimal places)?

ZScore_14C = ((12*5 + 11) - 64.6) / 2.52
print(f"If a woman is 5 feet 11 inches tall, then her Z-Score is {ZScore_14C}.")


# What percentage of women are taller than 5 feet 11 inches?  Round to 
# the nearest tenth of a percent.

from scipy.stats import norm
answer_14D = 1 - norm(loc = 64.6, scale = 2.52).cdf(12*5 + 11)
print(f"The proportion of women taller than 5 feet 11 inches is {answer_14D}.")
```
