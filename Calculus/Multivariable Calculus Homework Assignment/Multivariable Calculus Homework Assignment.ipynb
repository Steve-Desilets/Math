#### Answer

We know that the equation fo a tangent plane to a surface is given by the equation below.

z = f(x0, y0) + fx(x0, y0)(x - x0) + fy(x0, y0)(y - y0)

Accordingly, we will solve for each of the terms in the above equation and then piece them together to find our tangent plane equation

```python
from sympy import exp, symbols, diff

x, y = symbols('x, y')

f_x_y = 4*exp(-4*y + x**2)

partial_f_x = f_x_y.diff(x)

partial_f_y = f_x_y.diff(y)


def function_evaluator(func, X, Y):
    function_value = func.subs({x:X, y:Y})
    print(f"The value of f({X}, {Y}) is {function_value}.")

f_x0_y0 = f_x_y.subs({x:8, y:16})
fx_x0_y0 = partial_f_x.subs({x:8, y:16})
fy_x0_y0 = partial_f_y.subs({x:8, y:16})

z = f_x0_y0 + fx_x0_y0 * (x - 8) + fy_x0_y0 * (y - 16)
print(f"The equation for the plane tangent to z at (8, 16) is {z}.")

```

We will solve this problem using the equation for linear approximation of f(x, y) which is given below

### L(x, y) = f(x0, y0) + fx(x0,y0) * (x - x0) + fy(x0, y0) * ( y - y0)

```python
from sympy import symbols, Derivative

x, y = symbols('x, y')

f_x_y = 5 * ( (x*y) / 12) ** 0.5

partial_f_x = Derivative(f_x_y, x).doit()

partial_f_y = Derivative(f_x_y, y).doit()


def approx_function(X0, Y0, X1, Y1):
    f_x0_y0 = f_x_y.subs({x:X0, y:Y0})
    fx_x0_y0 = partial_f_x.subs({x:X0, y:Y0})
    X_diff = X1 - X0
    fy_x0_y0 = partial_f_y.subs({x:X0, y:Y0})
    Y_diff = Y1 - Y0
    Linear_Approx = f_x0_y0 + (fx_x0_y0 * X_diff) + (fy_x0_y0 * Y_diff)
    print(f"The linear approximation of (X1, Y1) is {Linear_Approx}.")

approx_function(8, 6, 8.12, 6.27)
```

For my second attempt, I'll use the Chain Rule for One Independent Variable to find the derivative of z with respect to t.

#### Chain Rule
#### dz/dt = (dz/dx) * (dx/dt) + (dz/dy) * (dy/dt)

```python
from sympy import Symbol, exp, Derivative

t = Symbol('t')

z_t = t**6 * exp(5 * (2 + 2*t))

dz_dt = Derivative(z_t, t).doit()

print(f"The derivative of z with respect to t is {dz_dt}.")
```

```python
from sympy import symbols, exp, Derivative

x, y, t = symbols('x, y, t')

z_x_y = x * exp(5*y)

dz_dx = Derivative(z_x_y, x).doit()

dz_dy = z_x_y.diff(y)

x_t = t**6

y_t = 2 + 2*t

dx_dt = Derivative(x_t, t).doit()

dy_dt = Derivative(y_t, t).doit()


dz_dt = dz_dx * dx_dt + dz_dy * dy_dt

print(f"The derivative of z with respect to t is {dz_dt}.")
```

#### Answer

We know that the chain rule states that 

#### dV/dt = dV/DT * dT/dt + dV/dP * dP/dt

The problem also gives us two of these values

#### dT/dt = 0.05

#### dP/dt = 0.02

We can also solve for dV/dT and dV/dP by solving for V.  

#### V = 8.31 * T / P

When we take the derivative of V with respect to P, we can treat T (which is 280) as a constant.  Similarly, when we take the derivative of V with respect to T, we can treat P (which is 19) as a constant.  This will enable us (hopefully) to find all the values that we need.

```python
from sympy import symbols

P, T, t = symbols('P, T, t')

V = (8.31 * T) / P

dV_dT = V.diff(T)

dT_dt = 0.05

dV_dP = V.diff(P)

dP_dt = 0.02

def derivative_calculator(temperature, pressure):
    dV_dT_value = dV_dT.subs({T:temperature, P:pressure})
    dV_dP_value = dV_dP.subs({T:temperature, P:pressure})
    dV_dt = dV_dT_value * dT_dt + dV_dP_value * dP_dt
    print(f"The rate at which the volume is changing is {dV_dt} L/s.")
    
    
derivative_calculator(280, 19)
```

#### Answer

Since this function is not subject to any constraints, we only need to check for maxima, minima, and saddle points by checking where the partial derivatives are zero or undefined.  We can determine whether each crtical point is a maximum, minimum, or saddle point using the discriminants.

```python
from sympy import symbols, solve

x, y = symbols('x, y')

f_x_y = x**3 + y**3 + 9*x**2 - 21*y**2 - 1

partial_f_x = f_x_y.diff(x)

partial_f_y = f_x_y.diff(y)

critical_x = solve(partial_f_x)

critical_y = solve(partial_f_y)

print(critical_x)
print(critical_y)
```

```python
from sympy import lambdify, Abs

def Discriminant_function(func, x_sym, y_sym, x_crit, y_crit):
    # Calculate the discriminant for a given function
    f_x_x = func.diff(x_sym, x_sym)
    f_y_y = func.diff(y_sym, y_sym)
    f_x_y = func.diff(x_sym, y_sym)
    
    # Create callable functions foro reach of the derivatives we created
    lambd_x_x = lambdify([x_sym, y_sym], f_x_x)
    lambd_y_y = lambdify([x_sym, y_sym], f_y_y)
    lambd_x_y = lambdify([x_sym, y_sym], f_x_y)
    
    fxx_ab = lambd_x_x(x_crit, y_crit)
    fyy_ab = lambd_y_y(x_crit, y_crit)
    fxy_ab = lambd_x_y(x_crit, y_crit)
    
    d = fxx_ab * fyy_ab - Abs(fxy_ab)**2
    print(f"The discriminant of our critical value ({x_crit}, {y_crit}) is {d}.")
    if d < 0:
          print(f"The point ({x_crit}, {y_crit}) is a saddle point")
    elif d == 0:
          print(f"The discriminant test for ({x_crit}, {y_crit}) is inconclusive.")
    elif d > 0:
          if fxx_ab < 0:
              print(f"The point ({x_crit}, {y_crit}) is a relative maxima")
          else:
              print(f"The point ({x_crit}, {y_crit}) is a relative minima")
          
          
Discriminant_function(f_x_y, x, y, -6, 0)
Discriminant_function(f_x_y, x, y, -6, 14)
Discriminant_function(f_x_y, x, y, 0, 0)
Discriminant_function(f_x_y, x, y, 0, 14 )
```

In this situation, we are aiming to minimize the cost function below subject to the constraint equation below

C(f, d, h) = 0.08fd + 2(0.03)dh + 0.03fh + 0.1fh

First we will find the critical points within the interior of the domain.

```python
from sympy import symbols, solve

f, d, h = symbols('f, d, h')
min_volume = 250
base_cost = 0.07
front_cost = 0.11
other_sides_cost = 0.04
C_f_d_h = base_cost*f*d + 2*other_sides_cost*d*h + other_sides_cost*f*h + front_cost*f*h

partial_C_f = C_f_d_h.diff(f)

partial_C_d = C_f_d_h.diff(d)

partial_C_h = C_f_d_h.diff(h)

solns = solve([
    partial_C_f,
    partial_C_d,
    partial_C_h
])

print(f"The critical point(s) occur at ({solns[f]}, {solns[d]}, {solns[h]}).")
```

Now that we've found the critical points within the interior of the domain, we must also leverage Lagrance Multipliers to find any maxima / minima points along the boundary of the domain.  

###  C(f, d, h) =   V(f, d, h)

The first equation above yields the following result:

###  (0.08 * d + 0.13 * h ) * i + ( 0.08 * f + 0.06 * h ) * j + ( 0.06 * d + 0.13 * f ) * k   =   (dhi + fhj + fdk)

Setting the coefficients of i, j, and k equal to one another yields the first three equations in our system of equations:

### 0.08d + 0.04h = 
### 0.08f + 0.06h = 
### 0.06d + 0.13f = 

The last equation in our system of equations (as defined by c(a,b) = 0) is:

### f * d * h - 300 = 0

We can solve this system of three equations to find maxima and minima along the boundary of the the domain of h.


```python
import sympy
l = symbols('l')
sympy.init_printing()
solns = sympy.solve([
    partial_C_f - l*d*h,
    partial_C_d - l*f*h,
    partial_C_h - l*f*d,
    f*d*h  - min_volume
])
```

Now that we have found our two critical points along the interior and boundary of the domain, we can plug these points back into our original function to determine which critical point results in the minimization of the cost function subject to our constraint that V => 300.
