```python
# Problem 5

# State the equations that you have as your inputs
# 6 = 2m + b
# -2 = 10m + b

# Solving each equation for b yields
# b = 6 - 2m
# b = -2 - 10m

# Setting these two equations equal to one another allows us to solve for m
# 6 - 2m = -2 - 10m
# 8m = -8
# m = -1

# We can now solve for b by plugging m into one of our original equations
# 6 = 2 * (-1) + b
# 8 = b

# So our equation in slope-intercept form is y = -x + 8
```

#### Answer

```python
# Problem 8

# First, determine the slope of profit changes over time using the information provided
def function_8a(y1, y2, x1, x2):
    return (y2 - y1) / (x2 - x1)

print(function_8a(39000, 123000, 1, 5))

# Second, determine the y-intercept of the desired equation
def function_8b(x, y, m):
    return y - (m * x)

print(function_8b(1, 39000, 21000))

# So the equation of our line of interest in slope-intercept form is y = (21000 * x) + 18000

# Use the newly defined function to predict the profit at the end of the 8th year
def function_8c(x):
    return (21000 * x) + 18000

print(function_8c(8))

# Predict when the company's profits will reach $417,000

def function_8d(y):
    return (y - 18000) / 21000

print(function_8d(417000))
```

#### Answer

```python
# Problem 14

# Import the packages that we will need for this exercise
import numpy as np
import matplotlib.pyplot as plt

# Create arrays to reflect the given data
x = np.array([65, 70, 75, 80, 85, 90, 95])
y = np.array([16.3, 18.1, 20.3, 22.5, 24.7, 27.2, 29.8])

#find line of best fit
a, b = np.polyfit(x, y, 1)

#add points to plot
plt.scatter(x, y, color='purple')

#add line of best fit to plot
plt.plot(x, a*x+b, color='steelblue', linestyle='--', linewidth=2)

#add fitted regression equation to plot
plt.text(1, 17, 'y = ' + '{:.3f}'.format(b) + ' + {:.3f}'.format(a) + 'x', size=14)
```

#### Answer

```python
# Problem 15

# Import the packages that we will need for this exercise
import numpy as np
import matplotlib.pyplot as plt

# Create arrays to reflect the given data
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([166, 195, 207, 229, 237, 292])

#find line of best fit
a, b = np.polyfit(x, y, 1)

#add points to plot
plt.scatter(x, y, color='purple')

#add line of best fit to plot
plt.plot(x, a*x+b, color='steelblue', linestyle='--', linewidth=2)

#add fitted regression equation to plot
plt.text(1, 17, 'y = ' + '{:.2f}'.format(b) + ' + {:.2f}'.format(a) + 'x', size=14)

# Use our best fit line (y = 22.23x + 165.43) to predict the amount of money that will be spent on football in 2026
def function_15(x):
    return (22.23 * x) + 165.43

print(function_15(2026 - 2008))

# Use the best fit line to predict the amount of money that will be spent on football in 2042
print(function_15(2042 - 2008))
```
